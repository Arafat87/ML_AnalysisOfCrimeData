```python
#Standard imports
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time

# machine learning library
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split

# machine learning accuracy
from sklearn.metrics import accuracy_score
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
df = pd.read_csv('drive/MyDrive/CSCI 355 Project Files/crimedata/crimedata_small.csv')
df
```

```python
df = df.dropna()
df.head()
```

```python
df.drop(['Block', 'IUCR', 'Description', 'Domestic', 'Latitude', 'Longitude'], axis=1, inplace=True)
df.head()
```

```python
df.drop(['Ward', 'FBI Code', 'X Coordinate', 'Y Coordinate', 'Updated On', 'Location'], axis=1, inplace=True)
df.drop(['Date', 'Case Number'], axis=1, inplace=True)
df.drop(['Location Description','Arrest'], axis=1, inplace=True)
df.head()
```

```python
print(df['Primary Type'].unique())
#print(df['Arrest'].unique())
print(df['District'].unique())
print(df['Community Area'].unique())
print(df['Year'].unique())
#Beat
#ID
```

```python
df.dropna(inplace=True)

df['Primary Type']= df['Primary Type'].map({'BATTERY':0, 'OTHER OFFENSE':1, 'ROBBERY':2, 'NARCOTICS':3, 'CRIMINAL DAMAGE':4,
 'WEAPONS VIOLATION':5, 'THEFT':6, 'BURGLARY':7, 'MOTOR VEHICLE THEFT':8,
 'PUBLIC PEACE VIOLATION':9, 'ASSAULT':10, 'CRIMINAL TRESPASS':11,
 'CRIM SEXUAL ASSAULT':12, 'INTERFERENCE WITH PUBLIC OFFICER':13, 'ARSON':14,
 'DECEPTIVE PRACTICE':15, 'LIQUOR LAW VIOLATION':16, 'KIDNAPPING':17, 'SEX OFFENSE':18,
 'OFFENSE INVOLVING CHILDREN':19, 'PROSTITUTION':20, 'HOMICIDE':21, 'GAMBLING':22,
 'INTIMIDATION':23, 'STALKING':24, 'OBSCENITY':25, 'PUBLIC INDECENCY':26,
 'HUMAN TRAFFICKING':27, 'CONCEALED CARRY LICENSE VIOLATION':28,
 'OTHER NARCOTIC VIOLATION':29, 'NON - CRIMINAL':30, 'NON-CRIMINAL':31,
 'NON-CRIMINAL (SUBJECT SPECIFIED)':32, 'RITUALISM':33}).astype(int)
#df['Arrest'] = df['Arrest'].map({'True':0, 'False':1}).astype(int, errors = 'ignore')

df.head()
```

```python
df['District']= df['District'].astype(int)
df['Community Area']= df['Community Area'].astype(int)
df.head()
```

```python
X = df.drop('Primary Type',axis=1)
y = df['Primary Type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42,shuffle=True)
```

```python
#Creating the KNN Model, use 20 neighbors given size of dataset for more accuracy

#Initialize Classifier
knn_model = KNeighborsClassifier(n_neighbors=3)

start = time.time()

#Train Model
knn_model.fit(X_train,y_train)

stop = time.time()

knn_time = stop - start

import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize) # change to false to re-truncate results

#Predict Model
y_pred = knn_model.predict(X_test)
y_pred

```

```python
print(accuracy_score(y_test, y_pred))
model_performance_knn = accuracy_score(y_test, y_pred)
```

```python
model_performance_destree = 0.0

  # reshape to get rid of the warning
  # not the most elegant, but it works
y_test  = np.reshape(y_test, len(y_test), )
y_train = np.reshape(y_train, len(y_train), )

model_tree = DecisionTreeClassifier()

start = time.time()

#Train Model
model_tree.fit(X_train, y_train)

stop = time.time()

tree_time = stop - start


y_pred = model_tree.predict(X_test)
model_performance_destree = accuracy_score(y_test, y_pred)
model_performance_destree
```

```python

model_performance_nb = 0.0

  # reshape to get rid of the warning
  # not the most elegant, but it works
y_test  = np.reshape(y_test, len(y_test), )
y_train = np.reshape(y_train, len(y_train), )

  # train the model
model_nb = GaussianNB()
model_nb.fit(X_train, y_train)

start = time.time()

#Train Model
model_nb.fit(X_train, y_train)

stop = time.time()

bayes_time = stop - start

  # test the model
y_pred = model_nb.predict(X_test)

  # accuracy results
model_performance_nb = accuracy_score(y_test, y_pred)
model_performance_nb
```

```python
result = pd.DataFrame({'Accuracy': [model_performance_knn, model_performance_svm, model_performance_destree, model_performance_nb],
                       'Time to Train:': [knn_time, svm_time, tree_time, bayes_time]})
result.index = ['KNN', 'SVM', 'Decesion Tree', 'Naive Bayes']
result
#Note: Accuracy is this low bc of the attributes relevance to the predicted crime
# Predicting Primary Type of Crime based on: ID	Beat	District	Community Area	Year
```
